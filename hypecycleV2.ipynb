## 2. Import of required libraries

```python
# Import numpy, pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib as matplot  # used to display version
import matplotlib.pyplot as plt
import math
import scipy.stats as stats

# Print library versions
print("Numpy installed version is: ", np.__version__, '          . Numpy is an open source library for scientific in Python.')
print("Pandas installed version is: ", pd.__version__, '         . Panda is an open source data analytis library in Python.' )
print("Matplotlib installed version is: ", matplot.__version__, '      . Matplotlib is an open graph generation library in Python.')
, 
# As a reference, current latest versions are (as of September 2020):
# Numpy: 1.19.1
# Panda: 1.1.1
# Matplotlib: 3.3.1 
```

## 3. Normal Distribution Curve 

```python
# The normal distribution represents the "hype portion" of the hypecyle. It is modelled using a normal distribution. 

mu = 2
variance = 2
sigma = math.sqrt(variance)
x = np.linspace(mu - 6*sigma, mu + 12*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.title('Normal Distribution (hype)')
plt.xlabel('Time')
plt.ylabel('Expectations')
plt.show()
```

## 4. S-Curve for Technology Maturity (Logistic function)

```python
# Dfinition de la fonction logistique
def flog(x,L,k,x0) :
    y = L / (1+np.exp(-k*(x-x0)))
    # Where:
    #       L represents the curve maximum value
    #       k is the logistic growth rate or steepness
    #       x0 is the value of the sigmoid mid point
    return y  

# Linespace definition
# Le premier paramtre est la premire valeur, le deuxime est la dernire valeur et le troisme, le nombre d'incrments
x = np.linspace(mu - 6*sigma, mu + 12*sigma, 100)

# Creation du graphique
fig, ax = plt.subplots() # Create a figure and an axes
plt.title('Technology S-Curve')
ax.plot(x, flog(x,0.15,0.3,4*mu), label='Technology maturity')

# Formatage du graphique
ax.set_xlabel("Time")
ax.set_ylabel("Performance/Benefit")
ax.legend()
```

## 5. Combining the Hype curve with the S-Curve = Hypecycle ! 

```python
x = np.linspace(mu - 6*sigma, mu + 12*sigma, 100)

y1 = stats.norm.pdf(x, mu, sigma) # Normal distribution
y2 = flog(x,0.15,0.3,4*mu)           # S-Curve 

fig, ax = plt.subplots() # Create a figure and an axes
ax.plot(x,y1+y2, label='Hypecycle')
#ax.plot(x,y1,label='Normal distribution')
#ax.plot(x,y2,label='Technology S-Curve')
plt.title('Hype Cycle')
plt.xlabel('Time')
plt.ylabel('Expectations')
ax.legend()
plt.show()
```
